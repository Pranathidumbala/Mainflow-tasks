```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=['turquoise','magenta'])
plt.title('Heart disease values')
plt.xlabel('1-heart disease,0-no heart disease')
plt.ylabel('amount');
```

```python
df.target.value_counts().plot(kind='pie',figsize=(7,5))
plt.legend(['disease','no disease']);
```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie',figsize=(7,5))
plt.title("male female ratio")
plt.legend(['male','female']);
```

```python
pd.crosstab(df.target,df.sex)
```

```python
df['sex']=df['sex'].map({0:'female',1:'male'})
sns.countplot(x='target',data=df,hue='sex')
plt.title('heart disease frequency for different genders')
plt.xlabel('0-No heart disease,1-heart disease')
plt.show()
```

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind = 'bar',color=['red','blue','violet','grey'])
plt.title('chest pain type vs count')
```

```python
pd.crosstab(df['cp'],df['target'])
```

```python
df['target']=df['target'].map({0:'no heaet disease',1:'heart disease'})
sns.countplot(x='cp',data=df,hue='target')
plt.title('heart disease frequency by chest pain type')
plt.xlabel('chest pain type (cp)')
plt.ylabel('frequency')
plt.show()
```
