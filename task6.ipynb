```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r'C:\Users\dpran\OneDrive\Desktop\diabetes_data.csv'
data = pd.read_csv(file_path)
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
data_statistics = data.describe()
print("\nBasic Statistics:\n", data_statistics)
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True, bins=20, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['BMI'], y=data['FastingBloodSugar'], hue=data['Diagnosis'], palette='coolwarm')
plt.title('Correlation Between BMI and Fasting Blood Sugar')
plt.xlabel('BMI')
plt.ylabel('Fasting Blood Sugar')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Diagnosis', data=data, palette='Set2')
plt.title('Distribution of Diabetes Diagnosis')
plt.xlabel('Diagnosis (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['PhysicalActivity'], y=data['BMI'], hue=data['Diagnosis'], palette='viridis')
plt.title('Physical Activity vs. BMI')
plt.xlabel('Physical Activity')
plt.ylabel('BMI')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='FamilyHistoryDiabetes', hue='Diagnosis', data=data, palette='muted')
plt.title('Family History of Diabetes and Diagnosis')
plt.xlabel('Family History of Diabetes')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diagnosis', y='HbA1c', data=data, palette='Set1')
plt.title('HbA1c Levels by Diabetes Diagnosis')
plt.xlabel('Diagnosis (0 = No, 1 = Yes)')
plt.ylabel('HbA1c')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Diagnosis', data=data, palette='cool')
plt.title('Gender-based Trends in Diabetes Diagnosis')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```
